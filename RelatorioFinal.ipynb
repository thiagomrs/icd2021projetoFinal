### **Dataset 1  - Brazilian Amazon Rainforest Degradation 1999-2019**

#### **Teste 1:**
**Hipótese nula:** a orientação política do presidente não afeta o desmatamento na Amazônia legal

**Hipótese alternativa:** a orientação política do presidente afeta o desmatamento na Amazônia legal

AMZ LEGAL: Tabela t: alpha = 0.05, upper-tail, df = 15 (graus de liberdade) ---> (1.7531)

Se o valor t encontrado (0.42445502467144486) fosse maior que o valor da tabela, rejeitaria a hi nula. Como não é o caso, a hi nula não pode ser rejeitada.

**Observação:** a inclusão dos dados de 2020 possivelmente alteraria o resultado, devido a "passagem da boiada"

## **5. Conclusão**

Um dos desafios enfrentados em análise de dados perpassa justamente na completude e corretude dos mesmos. Dados incompletos ou incorretos podem 
passar a sensação de ausência de correlação não necessariamente verdadeira. Em nossas análises nos deparamos com esse desafio, onde os datasets estudados não continham informações de todos os anos de todos os governos brasileiros (especialmente do último).
Embora seja de conhecimento público do crescimento expressivo do desmatamento e queimadas nos útltimos anos, não conseguimos provar a associação da ideologia governamental com o aumento da área desmatada.
